```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('/content/colleges_dataset_cleaned.csv')
```

```python
df.dtypes
```

```python
df.head()
```

```python
df_num = df.select_dtypes(include=['float64'])
```

```python
df_num.head()
```

```python
plt.scatter(df_num['graduation_rate'], df_num['acceptance_rate']) 
plt.show()
```

```python
fig = plt.figure(figsize=(6, 6))

gs = fig.add_gridspec(2, 2,  width_ratios=(4, 1), height_ratios=(1, 4),
                      left=0.1, right=0.9, bottom=0.1, top=0.9,
                      wspace=0.05, hspace=0.05)

scatter = fig.add_subplot(gs[1, 0])
histx = fig.add_subplot(gs[0, 0], sharex=scatter)
histy = fig.add_subplot(gs[1, 1], sharey=scatter)
histx.tick_params(axis="x", labelbottom=False)
histy.tick_params(axis="y", labelleft=False)

x, y = df_num['graduation_rate'].tolist(), df_num['acceptance_rate'].tolist()
# the scatter plot:
scatter.scatter(x, y)

# now determine nice limits by hand:
binwidth = 0.25
xymax = np.max([np.max(np.abs(x)), np.max(np.abs(y))])
lim = (int(xymax/binwidth) + 1) * binwidth

bins = np.arange(0, lim + binwidth, binwidth)
histx.hist(x, bins=bins)
histy.hist(y, bins=bins, orientation='horizontal')
plt.show()
```

```python
fig, axs = plt.subplots(2, 3)

axs[0, 0].boxplot(df_num['graduation_rate'])
axs[0, 0].set_title('Graduation Rate')

axs[0, 1].boxplot(df_num['acceptance_rate'])
axs[0, 1].set_title('Acceptance Rate')

axs[0, 2].boxplot(df_num['percent_of_student_who_get_grants'])
axs[0, 2].set_title('% of students who get grants')

axs[1, 0].boxplot(df_num['early_career_earnings'])
axs[1, 0].set_title("Early Career Earnings")

axs[1, 1].boxplot(df_num['avg_time_to_a_degree_years'])
axs[1, 1].set_title('Avg. time to a degree in years')

axs[1, 2].boxplot(df_num['median_sat'])
axs[1, 2].set_title('Median SAT')

fig.subplots_adjust(left=0.08, right=0.98, bottom=0.05, top=0.9,
                    hspace=1, wspace=2)

fig.figsize=(10,6)
plt.show()
```
